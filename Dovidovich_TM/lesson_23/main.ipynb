```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules, apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_csv('dataset.csv', header=None)
```

```python
data.head()
```

```python
data.fillna(' ', inplace=True)
```

```python
data = data.values.tolist()
```

```python
filtered_list = [list(filter(lambda x: x not in (None, ' '), row)) for row in data]
```

```python
te = TransactionEncoder()
new_data = te.fit(filtered_list).transform(filtered_list)
new_data = pd.DataFrame(new_data, columns=te.columns_)
```

```python
new_data
```

```python
sets = fpgrowth(new_data, min_support=0.01, use_colnames=True)
```

```python
sets
```

```python
rules = association_rules(sets)
```

```python
rules
```

```python
sets = apriori(new_data, min_support=0.01, use_colnames=True)
```

```python
sets
```

```python
rules = association_rules(sets)
```

```python
rules
```

В работе использовались 2 алгоритма: apriori и fpgrowth; в результе которых были построены ассоциативные правила. В правилах консеквентами в большинстве являются овощи. Оно и понятно, т.к. в 90 процентах покупок присутствуют овощи. Обработка датасета заключалась в избавлении от элементов таких как Nan и знака пробела. Dataframe был преобразован в двумерный список для построения ассоциативных правил.
