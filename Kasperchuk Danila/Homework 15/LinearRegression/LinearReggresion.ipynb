```python
import matplotlib_inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**2

def gr_func(x):
    return 2*x
```

```python
D = 5

X = np.linspace(-D, +D, 20)
Y = func(X)

plt.figure(figsize=(16,6))
plt.plot(X, Y, 'r', label='Y(X)')
```

```python
start_point = 5
D = 10

X = np.linspace(-D, +D, 20)
Y = func(X)

plt.figure(figsize=(16,6))
plt.plot(X, Y, 'r', label="Y(X)")
plt.plot(start_point, func(start_point), '-*g', label="GD")
```

```python
grad = gr_func(start_point)
grad
```
