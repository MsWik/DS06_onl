```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
from mlxtend.preprocessing import TransactionEncoder
```

Импорт нужных библиотек


```python
df = pd.read_csv('dataset.csv',header = None, na_values=' ')
df.head(7)
```

Просмотр датасета

```python
df.fillna(value="", inplace=True)
```

Залили все пустые значения на пустую строку

```python
transactions = []
for tr in range(0, df.shape[0]): 
    transactions.append([str(df.values[tr,j]) for j in range(0, df.shape[1])])
```

```python
encoder = TransactionEncoder()
encoder_ary = encoder.fit(transactions).transform(transactions)
df = pd.DataFrame(encoder_ary, columns=encoder.columns_)
df
```

```python
apr = apriori(df, min_support=0.1, use_colnames=True)
apr.sort_values(by=['support'], ascending = False)
```

Здесь мы использовали алгоритм априори, и затем отсортировали значения по максимальной поддержке

```python
rules = association_rules(apr, metric="lift", min_threshold=1)
rules.sort_values(by=['lift'], ascending=False)
```

Здесь мы использовали ассоциативные правила и также отсортировали по убыванию

```python
growth = fpgrowth(df, min_support=0.1, use_colnames=True)
growth.sort_values(by=['support'], ascending=False)
```

Использовали Frequent Pattern Growth, и отсортировали значения по поддержке от максимальной к минимальной

Выводы:
Я узнал, что алгоритмы из mlxtend.frequent_patterns работают действительно быстрее, и научился использовать их
