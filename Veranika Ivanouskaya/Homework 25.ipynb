#### построение ассоциативных правил для датасета 

```python
import pandas as pd
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
# Изменение датасета в список транзакций
transactions = []
for i in range(len(data)):
    transaction = data.iloc[i].dropna().tolist()
    transactions.append(transaction)
    
transactions
```

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
# Преобразование данных для one-hot encoding
encoder = TransactionEncoder()
ohe_array = encoder.fit(transactions).transform(transactions)
ohe_df = pd.DataFrame(ohe_array, columns=encoder.columns_)
```

```python
# Поиск частых наборов 
frequent_itemsets = apriori(ohe_df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
# Построение ассоциативных правил
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules
```

```python
# сортировка данных по уверенности и подъему
sorted_rules = rules.sort_values(by=['confidence', 'lift'], ascending=[False, False])
top_rules = sorted_rules.head(10)
top_rules # топ-10 правил по уверенности и подъему
```

В первой колонке можно увидеть наличие продуктов, которые часто покупают вместе. Овощи также присутствуют в  большинстве случаев. Все правила имеют значение подъема выше 1 - это значит, что antecedents и consequents встречаются вместе чаще, чем если бы они были независимы друг от друга. Среди antecedents видно разнообразие продуктов: стиральный порошок, йогурт, сливочное мороженое, яйца, бумажные полотенца и др. Это значит, что покупка овощей связана с широким спектром других продуктов, что указывает на тенденцию покупать все необходимое в одном месте
