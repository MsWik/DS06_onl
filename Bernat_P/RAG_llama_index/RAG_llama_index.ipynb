```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
system_prompt="""
You are expert in  Religion and
denominations in
the Republic of Belarus. Answer the question about  it
"""
## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="HuggingFaceH4/zephyr-7b-alpha",
    model_name="HuggingFaceH4/zephyr-7b-alpha",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16}
)
```

```python
model_name = 'all-MiniLM-L6-v2'
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model = HuggingFaceEmbedding(model_name=model_name)
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size = 1024,
    llm = llm,
    embed_model=embed_model
)
```

```python
from llama_index.core import SimpleDirectoryReader
```

```python
doc_pdf = SimpleDirectoryReader('/content/drive/My Drive/hw/data/').load_data()
```

```python
index_pdf = VectorStoreIndex.from_documents(doc_pdf,service_context= service_context)
```

```python
 query_engine_pdf= index_pdf.as_query_engine()
```

```python
query = '''what is
guaranteed by the Article 31
of the Constitution
of the Republic of Belarus?
'''
```

```python
answer = query_engine_pdf.query(query)
```

```python
print(answer)
```

According to the context information provided, the Article 31 of the Constitution of the Republic of Belarus guarantees the right to freedom of conscience and religion, which is regulated by the basic law of the Republic of Belarus and the law "On Freedom of Faith and Religious Organizations." This right includes the freedom to determine one's own attitude towards religion, to exercise any religion alone or together with others, or not to exercise any religion at all, to express and spread beliefs connected with one's attitude towards religion, to participate in worship, religious rituals, and rites, which are not prohibited by the law.
