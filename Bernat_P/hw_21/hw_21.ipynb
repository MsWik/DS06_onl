```python
import pandas as pd

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
bills = df.apply(lambda row: ','.join(row.values.astype(str)).split(','),axis=1)
```

```python
encoder = TransactionEncoder()
```

```python
transactions = pd.DataFrame(encoder.fit(bills).transform(bills),columns=encoder.columns_)
```

```python
del transactions[' ']
```

```python
transactions.columns
```

```python

display(transactions.head())
```

```python
frequent_itemsets = fpgrowth(transactions, min_support= 6/len(bills), use_colnames=True, max_len =3) 
```

```python
rules = association_rules(frequent_itemsets, metric="lift",  min_threshold = 1.5)
```

```python
display(rules.head(20))
```

```python
itemsets = apriori(transactions, 
                           use_colnames=True, 
                           verbose=1,
                           low_memory=False,
                           min_support=0.1
                          )

```

```python
itemsets.sort_values(by='support',ascending = False)
```

```python
rules = association_rules(itemsets, metric="lift", min_threshold=1)
rules.sort_values(by='confidence',ascending = False).head()
```
